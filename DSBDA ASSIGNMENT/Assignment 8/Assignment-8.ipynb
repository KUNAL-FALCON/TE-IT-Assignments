#  Line Chart


```python
sum_df = df.groupby(['Post Month'])['like'].sum()

#Making X Coordinates and Y coordinates
plt.plot(df['Post Month'].unique() , sum_df)

#Plotting Grapgh
plt.xlabel('Post Month')
plt.ylabel('Total Likes')
plt.title('Line Chart')

#Display Graph
plt.show()
```

#  Multiple Line Graph

```python
plt.plot(df['Post Month'].unique() , df.groupby(['Post Month'])['like'].sum())
plt.plot(df['Post Month'].unique() , df.groupby(['Post Month'])['comment'].sum())
plt.plot(df['Post Month'].unique() , df.groupby(['Post Month'])['share'].sum())

plt.ylabel('Total Likes')
plt.show()
```

```python
# get columns to plot
columns = ['Post Month' , 'Post Hour' , 'Post Weekday' , 'Paid']

# create x data
x_data = range(0, df.shape[0])

# create figure and axis
fig, ax = plt.subplots()

# plot each column
for column in columns:
    ax.plot(x_data, df[column], label=column)
    
# set title and legend
ax.set_title('FaceBook')
ax.legend()
```

```python
# get columns to plot
columns = df.columns.drop(['Type'])

# create x data
x_data = range(0, df.shape[0])

# create figure and axis
fig , ax = plt.subplots()

# plot each column
for column in columns:
    ax.plot(x_data, df[column], label = column)
    
# set title and legend
ax.set_title('FaceBook DataSet')
```
